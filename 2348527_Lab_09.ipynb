```python
import numpy as np
import matplotlib.pyplot as plt

# Functions for different activation types
def unit_step(x):
    return np.heaviside(x, 1)

def sign(x):
    return np.sign(x)

def linear(x, c=1):
    return c * x

def piecewise_linear(x):
    return np.piecewise(x, [x < 0, x >= 0], [lambda x: 0, lambda x: x])

def logistic(x):
    return 1 / (1 + np.exp(-x))

def hyperbolic_tangent(x):
    return np.tanh(x)

# Input values
input_values = np.linspace(5, 10, 100)  # Generate 100 input values from 5 to 10

# Calculate outputs for each activation function
unit_step_output = unit_step(input_values)
sign_output = sign(input_values)
linear_output = linear(input_values)
piecewise_linear_output = piecewise_linear(input_values)
logistic_output = logistic(input_values)
hyperbolic_tangent_output = hyperbolic_tangent(input_values)

# Plotting the graphs
plt.figure(figsize=(12, 8))

# Unit Step
plt.subplot(231)
plt.plot(input_values, unit_step_output)
plt.title('Unit Step Activation')
plt.xlabel('Input')
plt.ylabel('Output')

# Signum
plt.subplot(232)
plt.plot(input_values, sign_output)
plt.title('Signum Activation')
plt.xlabel('Input')
plt.ylabel('Output')

# Linear
plt.subplot(233)
plt.plot(input_values, linear_output)
plt.title('Linear Activation')
plt.xlabel('Input')
plt.ylabel('Output')

# Piece-wise Linear
plt.subplot(234)
plt.plot(input_values, piecewise_linear_output)
plt.title('Piece-wise Linear Activation')
plt.xlabel('Input')
plt.ylabel('Output')

# Logistic (Sigmoid)
plt.subplot(235)
plt.plot(input_values, logistic_output)
plt.title('Logistic Activation')
plt.xlabel('Input')
plt.ylabel('Output')

# Hyperbolic Tangent
plt.subplot(236)
plt.plot(input_values, hyperbolic_tangent_output)
plt.title('Hyperbolic Tangent Activation')
plt.xlabel('Input')
plt.ylabel('Output')

plt.tight_layout()
plt.show()
```

Assuming a simple neuron with a weighted input and a bias term, the output of the neuron can be calculated using the following formula:

Output = Activation_Function(∑(Wi * Xi) + b)

Where:

W: Weight
X: Input
b: Bias

Unit Step Activation Function: The unit step activation function produces a binary output of either 0 or 1. It can be defined as:
f(x) = 1 if x ≥ 0 f(x) = 0 if x < 0

We can create a graph that shows the step-like behavior of this function as the input value changes.

Sign (Signum) Activation Function: The sign activation function produces an output of -1 or 1 based on the sign of the input.
f(x) = -1 if x < 0 f(x) = 0 if x = 0 f(x) = 1 if x > 0

A graph can be created to display the signum nature of this function as the input changes.

Linear Activation Function: The linear activation function produces an output that is a linear function of the input value. It can be defined as:
f(x) = cx, where c is a constant value.

We can create a graph to display the linear relationship between the input and output values.

Piece-wise Linear Activation Function: The piece-wise linear activation function consists of multiple linear functions defined over different intervals.
We can create a graph to display the piece-wise linear nature of this function.

Logistic (Sigmoid) Activation Function: The logistic (sigmoid) activation function produces an output that is a sigmoidal function of the input value. It can be defined as:
f(x) = 1 / (1 + e^(-x))

We can create a graph to show the characteristic S-shaped curve of this function.

Hyperbolic Tangent Activation Function: The hyperbolic tangent activation function produces an output that is a hyperbolic tangent function of the input value. It can be defined as:
f(x) = tanh(x)

We can create a graph to display the hyperbolic tangent nature of this function.

```python
# Define the perceptron model
def perceptron_output(x1, x2, w1, w2, w0):
    net_input = w1 * x1 + w2 * x2 + w0
    return 1 if net_input >= 0 else 0
```

```python
# Define the weights
w1 = 0.5
w2 = 0.243
w0 = 0.2
```

```python
# Define the inputs for AND, OR, and XOR operations
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
```

```python
# Compute the output for AND operation
and_output = [perceptron_output(x1, x2, w1, w2, w0) for x1, x2 in inputs]
```

```python
# Compute the output for OR operation
or_output = [perceptron_output(x1, x2, w1, w2, w0) for x1, x2 in inputs]
```

```python
# Compute the output for XOR operation
xor_output = [perceptron_output(x1, x2, w1, w2, w0) for x1, x2 in inputs]
```

```python
# Define the expected outputs for AND, OR, and XOR operations
and_expected = [0, 0, 0, 1]
or_expected = [0, 1, 1, 1]
xor_expected = [0, 1, 1, 0]
```

```python
# Compute the error for each function
and_error = [abs(and_expected[i] - and_output[i]) for i in range(len(inputs))]
or_error = [abs(or_expected[i] - or_output[i]) for i in range(len(inputs))]
```

```python
# Display the outputs and errors
print("AND Operation - Output:", and_output, "Error:", and_error)
print("OR Operation - Output:", or_output, "Error:", or_error)
```
